#ðŸ”§ TransformaciÃ³n

```python
# se normaliza las columnas correspondientes
customer_data = pd.json_normalize(datos_telecom['customer'])
phone_data = pd.json_normalize(datos_telecom['phone'])
internet_data = pd.json_normalize(datos_telecom['internet'])
account_data = pd.json_normalize(datos_telecom['account'])

# Concatenando el nuevo dataframe con el original dataframe
datos_telecom_expanded = pd.concat([datos_telecom[['customerID', 'Churn']], customer_data, phone_data, internet_data, account_data], axis=1)

display(datos_telecom_expanded.head())
```

```python
# Revision del tipo de datos de cada columna y su consistencia con sus datos
datos_telecom_expanded.info()
```

```python
# cambio a numerico tipo float64, numero con decimales
 # errors='coerce' se asegurÃ³ de que cualquier valor en esa columna que no pudiera ser convertido a nÃºmero
  # (como cadenas vacÃ­as o texto) fuera reemplazado por NaN (Not a Number), que representa valores faltantes.
datos_telecom_expanded['Charges.Total'] = pd.to_numeric(datos_telecom_expanded['Charges.Total'], errors='coerce')
print(datos_telecom_expanded.dtypes)
```

```python
datos_telecom_expanded.info()
```

```python
# RevisiÃ³n de Porcentajes
# Agrupar por tenure y Churn
df_porcentaje = datos_telecom_expanded.groupby(['tenure', 'Churn']).size().unstack(fill_value=0)
```

```python
# Calcular el % de churn por tenure
df_porcentaje['churn_rate'] = (df_porcentaje.get('Yes', 0) / df_porcentaje.sum(axis = 1)) * 100

# Mostrar el resultado
print(df_porcentaje[['churn_rate']])
```

```python
# Graficamos los porcentajes
fig = px.line(df_porcentaje.reset_index(),
              x='tenure',
              y='churn_rate',
              markers = True,
              title='Porcentaje de Churn (Clientes Retirados) por Tenure (Meses de Contrato)')
fig.show()
```
