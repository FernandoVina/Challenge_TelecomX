```python
# se normaliza las columnas correspondientes
customer_data = pd.json_normalize(datos_telecom['customer'])
phone_data = pd.json_normalize(datos_telecom['phone'])
internet_data = pd.json_normalize(datos_telecom['internet'])
account_data = pd.json_normalize(datos_telecom['account'])

# Concatenando el nuevo dataframe con el original dataframe
datos_telecom_expanded = pd.concat([datos_telecom[['customerID', 'Churn']], customer_data, phone_data, internet_data, account_data], axis=1)

display(datos_telecom_expanded.head())
```

```python
datos_telecom_expanded.info()
```

```python
datos_telecom_expanded['Charges.Total'] = pd.to_numeric(datos_telecom_expanded['Charges.Total'], errors='coerce')
# errors='coerce' se asegurÃ³ de que cualquier valor en esa columna que no pudiera ser convertido a nÃºmero
 # (como cadenas vacÃ­as o texto) fuera reemplazado por NaN (Not a Number), que representa valores faltantes.
```

```python
print(datos_retirados.shape[0])
```

```python
total_monthly_charges_all = datos_telecom_expanded['Charges.Monthly'].sum()
total_total_charges_all = datos_telecom_expanded['Charges.Total'].sum()

print(f"Total Monthly Charges for all customers: ${total_monthly_charges_all:.2f}")
print(f"Total Total Charges for all customers: ${total_total_charges_all:.2f}")
```

```python
total_monthly_charges_churned = datos_retirados['Charges.Monthly'].sum()
total_total_charges_churned = datos_retirados['Charges.Total'].sum()

print(f"Total Monthly Charges for churned customers: ${total_monthly_charges_churned:.2f}")
print(f"Total Total Charges for churned customers: ${total_total_charges_churned:.2f}")
```

```python
categorical_columns = datos_telecom_expanded.select_dtypes(include='object').columns
columns_to_exclude = ['customerID', 'Churn']
# Exclude 'customerID' and 'Churn' as it will always be 'Yes' in datos_retirados

for col in categorical_columns:
    if col not in columns_to_exclude:
        print(f"Comparing '{col}':")
        print("Full Dataset:")
        display(datos_telecom_expanded[col].value_counts())
        print("\nChurned Customers Dataset:")
        display(datos_retirados[col].value_counts())
        print("-" * 30)
```
