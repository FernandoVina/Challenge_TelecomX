```python
customer_data = pd.json_normalize(datos_telecom['customer'])
phone_data = pd.json_normalize(datos_telecom['phone'])
internet_data = pd.json_normalize(datos_telecom['internet'])
account_data = pd.json_normalize(datos_telecom['account'])

# Concatenando el nuevo dataframe con el original dataframe
datos_telecom_expanded = pd.concat([datos_telecom[['customerID', 'Churn']], customer_data, phone_data, internet_data, account_data], axis=1)

display(datos_telecom_expanded.head())
```

```python
datos_telecom_expanded['Charges.Total'] = pd.to_numeric(datos_telecom_expanded['Charges.Total'], errors='coerce')
```
